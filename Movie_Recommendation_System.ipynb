```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv("tmdb_5000_movies.csv")
credits = pd.read_csv("tmdb_5000_credits.csv")
movies.head(2)
```

```python
credits.head(2)
```

```python
# credits.head(1)['cast'].values
# movies.head(1)['genres'].values
```

```python
# now we have two dataframes , we will merge these two dataframe on basis of similar column
movies = movies.merge(credits,on='title')
movies.head(1)
```

```python
# columns which we need to create tags


# movie_id
# title
# overview
# genres
# keywords
# cast
# crew
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
movies.head()
```

```python
movies.isnull().sum()  #check foe the missing data in any column
```

```python
# we have 3 missing column in overview column , so just drop the column
movies.dropna(inplace = True)
movies.isnull().sum()
```

```python
# check for duplicate data
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
# we are getting the object which is list of dictionary but in string format
# '[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'
# desired format :  ['Action','Adventure','Fantasy','Science Fiction']
```

```python
import ast    
```

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        tag = i['name']
        L.append(tag)
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

```python
# we can use the same convert function to clean the column 'keywords'
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
# movies.iloc[0].cast
# we can use the same convert function to clean the column 'cast'
movies['cast'] = movies['cast'].apply(convert)
movies.head()
```

```python
# now we will clean the column crew
# we only fetch the name which contain the job = director
```

```python
def fetch_director(text):
    List = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            List.append(i['name'])
    return List
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
movies.head()
```

```python
# now we have to clean the column 'overview' , convert the string into list of elements
movies['overview'] = movies['overview'].apply(lambda x:x.split())
movies.head()
```

```python
# now we have to remove whitespace of tags, if user type sam then our model will might confuse which sam to show
# Sam Worthington => SamWorthington , Sam Mendes => SamMendes 
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])

movies.head()
```

```python
movies['tags'] = movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']  # bag of words
```

```python
movies.head()
```

```python
new_df = movies[['movie_id','title','tags']]
new_df
```

```python
new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
import nltk

from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():              # text.split() converts the text into list
        y.append(ps.stem(i))

    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
 new_df['tags'][0]
```

```python
# now we will try to make vectors of all the tags in tags column
# we will use library sklearn and import CountVectorizer
# max_feature=5000  this indicates how many similar words we want , stop_words='english' -> ignores the english stop words
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')  # we will get 5000 most similar words from all the rows of tags column
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
# here we are converting it to numpy array because sklearn returns scipi sparse matrix
```

```python
vectors.shape
```

```python
# we have 4806 movies and 5000 vectors
```

```python
vectors[0]
```

```python


# printing 100 features out of 5000
feature_names = cv.get_feature_names_out()
for feature in feature_names[:100]:
    print(feature)
```

```python
# now we can see there are various similar features like accept , accepting , accepted etc 
# so we apply stemming of words => [accept , accepting , accepted] => after stemming => [accept, accept, accept]
# we will perform this operation above line 32   and re run the other steps after it
```

```python
# now we will calculate cosine distance of each vector with every other vector of movies to calculate similarity
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors)
```

```python
cosine_similarity(vectors).shape
```

```python
# total number of comparisions are 4806 x 4806
```

```python
similarity = cosine_similarity(vectors)
similarity
```

```python
similarity[0]   # it shows the similarity of first movie with all other 4806 movies 
```

```python
# now we have to create a function which when user gives any one movie name then it gives or recomment the 5 similar movies 
```

```python
# to find the index of any movie 
movie_index = new_df[new_df['title']== 'Batman Begins'].index[0]
movie_index
```

```python
# defining index of each comparision of similarity matrices and sorting it on basis of comparision instead of index
sorted(list(enumerate(similarity[0])), reverse=True, key = lambda x:x[1])[1:6]
```

```python
# creating a function
def recommend(movie):
    movie_index = new_df[new_df['title']== movie].index[0]
    # distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(similarity[movie_index])), reverse=True, key = lambda x:x[1])[1:6]

    for i in movies_list:
        # print (i[0])
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
